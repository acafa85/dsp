```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mi_funcion_sen(vmax, dc, ff, ph, nn, fs):
    '''
    : Parametros:  
        vmax : Amplitud [V]
        dc   : Valor medio [V]
        ff   : Frecuencia de la señal [Hz]
        ph   : fase [radianes]
        nn   : cantidad de muestras digitalizada por el ADC (# muestras)
        fs   : Frecuencia de muestreo [Hz]  
    '''
    
    Ts = 1/fs
    t = nn * Ts # Longitud de tiempo de la senoidal segun muestras y Periodo de muestreo
    x = np.linspace(0, (nn-1)*Ts, nn).flatten()
    nn = np.arange(nn)
    y = vmax*np.sin(2*np.pi*ff*nn*Ts + ph*(np.pi/180)) + dc
    return x,y
```

```python
N = 300
fs = 5000
tt, xx = mi_funcion_sen(vmax = 3, dc = 3, ff = 50, ph=0, nn = N, fs = fs)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.grid()
plt.plot(tt,xx)
```
