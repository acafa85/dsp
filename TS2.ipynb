```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-poster')
%matplotlib inline
```

DFT : X = $\sum_{n=0}^{N-1}{x_n\cdot e^{-i2\pi{kn/N}}} = \sum_{n=0}^{N-1}{x_n[cos(2\pi{kn/N}) -i\cdot sin(2\pi{kn/N})]}$

```python
def mi_funcion_sen(vmax, dc, ff, ph, nn, fs):
    '''
    : Parametros:  
        vmax : Amplitud [V]
        dc   : Valor medio [V]
        ff   : Frecuencia de la se√±al [Hz]
        ph   : fase [radianes]
        nn   : cantidad de muestras digitalizada por el ADC (# muestras)
        fs   : Frecuencia de muestreo [Hz]  
    '''
    
    Ts = 1/fs
    t = nn * Ts # Longitud de tiempo de la senoidal segun muestras y Periodo de muestreo
    x = np.linspace(0, (nn-1)*Ts, nn).flatten()
    nn = np.arange(nn)
    y = vmax*np.sin(2*np.pi*ff*nn*Ts + ph*(np.pi/180)) + dc
    return x,y
```

```python
N = 300
fs = 5000
tt, xx = mi_funcion_sen(vmax = 3, dc = 3, ff = 50, ph=0, nn = N, fs = fs)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.grid()
plt.plot(tt,xx)
```

amp = $\frac{|X_k|}{N}= \frac{\sqrt{Re(X_k)^2 + Im(X_k)^2}}{N}$

phase = $atan2(Im(X_k), Re(X_k))$

```python
def DFT(x):
    """
    Funcion para calcular la Transformada Discreta de Fourirer
    de 1 dimension para una se√±al x
    """
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    # parte exponencial:
    e = np.exp(-2j * np.pi * k * n / N)
    # np.dot hace el producto interno
    X = np.dot(e, x)
    
    return X
```

```python
X = DFT(xx)

# calculate the frequency
N = len(X)
n = np.arange(N)
T = N/fs
freq = n/T 

plt.figure(figsize = (8, 6))
plt.stem(freq, abs(X), 'b')
plt.xlabel('f [Hz]')
plt.ylabel('Amplitud DFT |X(freq)|')
plt.show()
```

```python
n_oneside = N//2
# get the one side frequency
f_oneside = freq[:n_oneside]

# normalize the amplitude
X_oneside =X[:n_oneside]/n_oneside

plt.figure(figsize = (12, 6))

plt.ylabel('DFT Amplitude |X(freq)|')

plt.subplot(122)
plt.stem(f_oneside, abs(X_oneside), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```
