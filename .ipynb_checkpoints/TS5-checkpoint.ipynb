```python
# -*- coding: utf-8 -*-
"""
Created on Wed Sep 15 22:12:36 2021

@author: ADRIAN CAFA
"""
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
import scipy.stats as stats
import matplotlib.patches as mpatches
from scipy.fft import *
mpl.rcParams['figure.figsize'] = (12,13)
```

```python
plt.close("all")

N = 1000 # numero de muestras
fs= 1000 # frecuencia de sampling Hz

ts = 1/fs # tiempo de muestreo

t=np.arange(0,1,ts)

# Del enunciado solicitado:

K0 = [N/4, N/4+0.25, N/4+0.5]

fase=0 
df=fs/N
amp=1 #amplitud en volts

f=np.arange(0,fs,df)
bfrec = f <= fs/2
```

```python
# K0: seno_0

freq0=K0[0]*fs/N # frecuencia en Hz
seno_0 = np.sqrt(amp)*np.sin(2 * np.pi * freq0 * t + fase)
```

```python
# K1: seno_1

freq1=K0[1]*fs/N # frecuencia en Hz
seno_1 = np.sqrt(amp)*np.sin(2 * np.pi * freq1 * t + fase)
```

```python
# K2: seno_2

freq2=K0[2]*fs/N # frecuencia en Hz
seno_2 = np.sqrt(amp)*np.sin(2 * np.pi * freq2 * t + fase)
```

```python
# El Valor de la energia en el tiempo debe ser constante en frecuencia por Parseval 

E0=1/N*sum(np.abs(seno_0)**2)
E1=1/N*sum(np.abs(seno_1)**2)
E2=1/N*sum(np.abs(seno_2)**2)


print("POTENCIA seno_0: ",E0)
print("POTENCIA seno_1: ",E1)
print("POTENCIA seno_2: ",E2)
```

```python
# Normalizamos la señal:

sig_norm0 = seno_0 / np.sqrt(E0)
sig_norm1 = seno_1 / np.sqrt(E1)
sig_norm2 = seno_2 / np.sqrt(E2)


print("Potencia=",np.var(sig_norm0))
print("Potencia=",np.var(sig_norm1))
print("Potencia=",np.var(sig_norm2))
```

```python
# Calculo la densidad espectral de potencia
# Al integrar las funciones densisdad de potencia deberian ser iguales a sus correspondientes
# Energias en tiempo

DEP_0 = np.abs(1/N*fft(sig_norm0))**2
DEP_1 = np.abs(1/N*fft(sig_norm1))**2
DEP_2 = np.abs(1/N*fft(sig_norm2))**2

DEP_0_int = sum(DEP_0)
DEP_1_int = sum(DEP_1)
DEP_2_int = sum(DEP_2)


print("Pot en tiempo=",E0,"Potencia normalizada en frec=",DEP_0_int,"para señal 0")
print("Pot en tiempo=",E1,"Potencia normalizada en frec=",DEP_1_int,"para señal 1")
print("Pot en tiempo=",E2,"Potencia normalizada en frec=",DEP_2_int,"para señal 2")
```

```python
# Visualizacion de las DEP:
# Nota: Al muestrear sobre la mitad de las frecuencias tengo que hacer el doble la DEP

plt.figure(1)
# Hago el doble la DEP porque es funcion par el modulo

plt.figure(1)
plt.stem(f,2*DEP_0)
plt.title("DEP_0 - k0 = {:3.2f}".format(K0[0]))
plt.ylabel("Potencia[W]")
plt.xlabel("frec[Hz]")
plt.xlim(245,255)

plt.figure(2)
plt.stem(f,2*DEP_1)
plt.title("DEP_1 - k0 = {:3.2f}".format(K0[1]))
plt.ylabel("Potencia[W]")
plt.xlabel("frec[Hz]")
plt.xlim(245,255)

plt.figure(3)
plt.stem(f,2*DEP_2)
plt.title("DEP_2 - k0 = {:3.2f}".format(K0[2]))
plt.ylabel("Potencia[W]")
plt.xlabel("frec[Hz]")
plt.xlim(245,255)

plt.show()
```

En el caso inicial,donde no tenemos desintonía se ve la delta situada en 250 Hz con potencia unitaria. En el caso donde la desintonía es la más grande de todas (tercer caso) se verifico un desparrame espectral mayor y alrededor de los 250 Hz. Como mi resolucion espectral no vario para las 3 señales, entonces puede llegar a ajustarse mejor a esa resolucion algunas veces pero se verifico que no sucede siempre.

### Técnica de Zero padding

Esta técnica consiste en agregar ceros al final de la señal para aumentar Δf (de forma ficticia) lo cual permite contar con mayor precisión para visualizar el desparramo espectral de potencia visto anteriormente por ejemplo.

```python
# Se agregará un vector de 9*N ceros al final (Script visto en clase)


Npad = N*10
f_pad = np.linspace(0, (Npad-1)*(fs/(Npad)), Npad)
t_pad = np.linspace(0, (Npad-1)/fs, Npad)

xx = np.hstack ((
    np.sin(2*np.pi*K0[0]*(fs/N)*t).reshape(N,1),
    np.sin(2*np.pi*K0[1]*(fs/N)*t).reshape(N,1),
    np.sin(2*np.pi*K0[2]*(fs/N)*t).reshape(N,1)
    ))

zz = np.zeros_like(xx)


# Nueva señal
xx_pad = np.vstack([xx,zz.repeat(9,axis=0)])

# normalizado por potencia:
xx_pad[:,0] = xx_pad[:,0]/np.sqrt(np.mean(xx_pad[:,0]**2,axis=0))
xx_pad[:,1] = xx_pad[:,1]/np.sqrt(np.mean(xx_pad[:,1]**2,axis=0))
xx_pad[:,2] = xx_pad[:,2]/np.sqrt(np.mean(xx_pad[:,2]**2,axis=0))

plt.figure(5)

plt.plot(t_pad,xx_pad[:,0])
plt.title("Señal zero padding")
plt.ylabel("Magnitud")
plt.xlabel("segundos")

XX_pad = np.zeros_like(xx_pad)
XX_pad[:,0] = np.abs((1/(Npad))*fft(xx_pad[:,0]))**2
XX_pad[:,1] = np.abs((1/(Npad))*fft(xx_pad[:,1]))**2
XX_pad[:,2] = np.abs((1/(Npad))*fft(xx_pad[:,2]))**2
```

```python
# Grafico en dB:

DEP0_pad = XX_pad[:,0]
DEP1_pad = XX_pad[:,1]
DEP2_pad = XX_pad[:,2]

plt.figure(6)
plt.plot( f_pad, 10* np.log10(2*DEP0_pad),':x', lw=2, label='$ k0 = $' + '{:3.2f}'.format(K0[0]))
plt.plot( f_pad, 10* np.log10(2*DEP1_pad),':x', lw=2, label='$ k0 = $' + '{:3.2f}'.format(K0[1]))
plt.plot( f_pad, 10* np.log10(2*DEP2_pad),':x', lw=2, label='$ k0 = $' + '{:3.2f}'.format(K0[2]))
plt.title("DSP [dB] con zero padding")
plt.ylabel("DSP [dB]")
plt.xlabel("frec[Hz]")
plt.xlim(220,280)
axes_hdl = plt.gca()
axes_hdl.legend()
plt.show()

plt.figure(7)
plt.plot( f_pad, 10* np.log10(2*DEP0_pad),':x', lw=2, label='$ k0 = $' + '{:3.2f}'.format(K0[0]))
plt.plot( f_pad, 10* np.log10(2*DEP1_pad),':x', lw=2, label='$ k0 = $' + '{:3.2f}'.format(K0[1]))
plt.plot( f_pad, 10* np.log10(2*DEP2_pad),':x', lw=2, label='$ k0 = $' + '{:3.2f}'.format(K0[2]))
plt.title("DSP [dB] con zero padding con zoom")
plt.ylabel("Densidad de Potencia [dB]")
plt.xlabel("frec[Hz]")
plt.xlim(245,255)
plt.ylim(-50,5)
axes_hdl = plt.gca()
axes_hdl.legend()

plt.show()
```

Con zero padding obtenemos una resolución espectral mas fina ya que nos permite interpolar los valores de frecuencia de manera más precisa que sin su uso.
