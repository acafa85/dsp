```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy.signal as sig
from scipy.fft import fft, fftshift
import matplotlib.patches as mpatches
from pandas import DataFrame
from IPython.display import HTML
import spectrum
from spectrum.datasets import marple_data
from pylab import legend, ylim
norm = True
sides = 'centerdc'
mpl.rcParams['figure.figsize'] = (10,10)
```

```python
#######################################################################################################################
#%% Inicio de la simulación
#######################################################################################################################
plt.close('all')

# Datos generales de la simulación
fs = 1000 # frecuencia de muestreo (Hz)
N = 1000   # cantidad de muestras

figura=0
 
ts = 1/fs # tiempo de muestreo
df = fs/N # resolución espectral

t=np.arange(0,1,ts)
f=np.arange(0,fs,df)
####################
###Armo la señal
####################
a1=1
# SNR=10*log(a1/noise_power)
SNR=np.array([3,10])
noise_power=np.zeros_like(SNR)
noise=np.zeros_like(SNR)
noise_power=a1/(10**(SNR/10))

print(noise)
print(noise_power)

omega_0=np.pi/2
fr=np.random.uniform(low=-1/2, high=1/2, size=200)

omega_1=omega_0+fr*2*np.pi/N

print(omega_1)
```

```python
##cambio ejes para que sea (200,1)*(1,1000)
####################
###1) Experimento con 3dB
####################
noise_0=np.random.normal(loc=0,scale=np.sqrt(noise_power[0]),size=(1000,200))
senal=np.sqrt(2*a1)*np.sin(2*np.pi*omega_1.reshape(1,200)*(fs/(2*np.pi))*t.reshape(1000,1))
x_0=senal + noise_0
```

```python
####################
###2) Experimento con 10dB
noise_1=np.random.normal(loc=0,scale=np.sqrt(noise_power[1]),size=(1000,200))
x_1=senal + noise_1
```

```python
####################################
### Periodograma con 3 dB
####################################

#calculo el periodograma normal
Pp=(1/N)*np.abs((fft(x_0,axis=0)))**2

#Metodo de Welch
f_welch,Pxx_den = sig.welch(x_0,fs=fs,nperseg=N/5,axis=0)


##Grafico las salidas
plt.figure(figura)
figura+=1

# Zoom:
plt.plot(f,Pp)
plt.xlim(240,260)

plt.figure(figura)
figura+=1
#plt.semilogy(f_welch,Pxx_den)
plt.plot(f_welch,Pxx_den)
plt.xlim(240,260)

plt.figure(figura)
figura+=1
for i in range(len(fr)):
    p_BT = spectrum.pcorrelogram(x_0[:,i], lag=15)
    p_BT(); 
    p_BT.plot(label='Correlogram(15)', norm=norm, sides=sides)
plt.xlim(0.1,0.4)
plt.ylim(-20,1)
```

```python
################################
###Calculo del pico con 3dB
################################

indices_Pp_0=Pp.argmax(axis=0)
indices_Pxx_den_0=Pxx_den.argmax(axis=0)

################################
### Calculo el error
### Con los dos metodos para 3dB
################################

error_Pp_0=f[indices_Pp_0]-(omega_1*fs/(2*np.pi))
error_Pxx_0=f_welch[indices_Pxx_den_0]-(omega_1*fs/(2*np.pi))
media_Pp_0=np.mean(error_Pp_0)
media_Pxx_0=np.mean(error_Pxx_0)

################################
### Calculo la varianza
### Con los dos metodos para 3dB
################################

var_error_Pp_0=np.var(error_Pp_0)
var_error_Pxx_0=np.var(error_Pxx_0)
```

```python
################################
###Calculo del periodograma con 10 dB
################################

#calculo el periodograma normal
Pp=(1/N)*np.abs((fft(x_1,axis=0)))**2

###############################################################
######################Metodo de Welch##########################
###############################################################
f_welch,Pxx_den = sig.welch(x_1,fs=fs,nperseg=N/5,axis=0)

# Output:
plt.figure(figura)
figura+=1
##En este grafico restrinjo el eje x ya que no queremos la otra mitad
plt.plot(f,Pp)
plt.xlim(240,260)


plt.figure(figura)
figura+=1
#plt.semilogy(f_welch,Pxx_den)
plt.plot(f_welch,Pxx_den)
plt.xlim(240,260)

############
#Spectrum no tiene ninguna implementacion para axis, entonces armo las comparaciones con 1 muestra a la vez
############
plt.figure(figura)
figura+=1
for i in range(len(fr)):
    p_BT = spectrum.pcorrelogram(x_1[:,i], lag=15)
    p_BT(); 
    p_BT.plot(label='Correlogram(15)', norm=norm, sides=sides)
plt.xlim(0.1,0.4)
plt.ylim(-20,1)
```

```python
################################
###Calculo del pico con 10dB
################################

indices_Pp_1=Pp.argmax(axis=0)
indices_Pxx_den_1=Pxx_den.argmax(axis=0)

#######################################
### Error con los dos metodos para 10dB
#######################################
# Periodograma
error_Pp_1=f[indices_Pp_1]-(omega_1*fs/(2*np.pi))

# Welch
error_Pxx_1=f_welch[indices_Pxx_den_1]-(omega_1*fs/(2*np.pi))

# Medias:
media_Pp_1=np.mean(error_Pp_1)
media_Pxx_1=np.mean(error_Pxx_1)

################################
### Calculo la varianza
### Con los dos metodos para 10dB
################################

var_error_Pp_1=np.var(error_Pp_1)
var_error_Pxx_1=np.var(error_Pxx_1)
```

```python
#######################
#%%Calculo de sesgo y varianza
#######################

#Sesgo: s = mu - a0

s_Periodo_3dB = media_Pp_0
s_Welch_3dB = media_Pxx_0

s_Periodo_10dB = media_Pp_1
s_Welch_10dB = media_Pxx_1


data = [[s_Periodo_3dB, var_error_Pp_0],
        [s_Periodo_10dB, var_error_Pp_1],
        [s_Welch_3dB, var_error_Pxx_0],
        [s_Welch_10dB, var_error_Pxx_1]]

df = DataFrame(data,columns=['$s_a$', '$v_a$'],
                index=[  
                        'Periodograma 3dB',
                        'Periodograma 10dB',
                        'Welch 3dB',
                        'Welch 10dB'
                      ])
HTML(df.to_html())

```

### Conclusiones:

El periodograma es un estimador de la densidad espectral de potencia (PSD) Φ𝑥𝑥(ejΩ) de una señal aleatoria 𝑥[𝑘] . 
Se asume que la señal 𝑥[𝑘] se extrae de un proceso aleatorio de valor real ergódico. (Un sistema es ergódico si su valor esperado, es decir el promedio de muchos sistemas independientes realizando el experimento, es igual a su promedio a largo plazo).
El periodograma como estimador no paramétrico de la densidad espectral de potencia (PSD) Φ𝑥𝑥(ejΩ) de una señal aleatoria 𝑥[𝑘] no es consistente. Esto se debe al hecho de que su varianza no converge hacia cero incluso cuando la longitud de la señal aleatoria se incrementa hacia el infinito. Para superar este problema, el método de Bartlett y el método de Welch dividen la señal aleatoria en segmentos, estiman la PSD de cada segmento y promedian estas estimaciones localmente.
El promedio reduce la varianza de la PSD estimada. Mientras que el método de Barlett utiliza segmentos no solapados, el de Welch es una generalización que utiliza segmentos solapados en ventanas. Para el método de Welch se asume un proceso aleatorio de valor real ergódico en sentido amplio.
Al visualizar las señales se verifica que con el metodo de welch es mejor para poder definir en que frecuencia esta nuestra señal comparado con un periodograma normal que dispersa mucho mas la energia.
